```python
import re
import numpy as np
import pandas as pd
import csv
```

```python
def extract_country_names(text):
    pattern = r"(?<=\]).*?(?=\[|$)"  # 匹配方括号内的内容

    matches = re.findall(pattern, text)  # 使用正则表达式查找所有匹配的内容

    # 提取国家名字
    countries = []
    for match in matches:
        parts = match.split(';')  # 将匹配到的内容按分号分割成多个部分
        for part in parts:
            if ',' in part:  # 确保含有逗号
                country = part.split(',')[-1].strip()  # 获取逗号后面的部分，并去除首尾空格
                countries.append(country)

    return countries
```

```python
def writer_nums(text):
    pattern = r"\[(.*?)\]"  # 匹配方括号内的内容

    matches = re.findall(pattern, text)  # 使用正则表达式查找所有匹配的内容
    writers=[]
    for match in matches:
        parts = match.split(';')  # 将匹配到的内容按分号分割成多个部分
        for part in parts:
            writers.append(part)
    return writers
```
